# ◆NRC 감성사전
* 등록된 14,182개의 각 단어들에 대해 긍정, 부정 + 8가지의 감정 총 10가지 척도에 대해 0 또는 1로 라벨링되어있다.

```python
nrc = pd.read_csv('./data_files/NRC.txt', engine='python', header=None, sep='\t')
```

```python
nrc
```

```python
nrc.iloc[:,1].value_counts()   # 감성 종류
```

```python
nrc.iloc[:,0].value_counts()
```

```python
nrc = nrc[(nrc != 0).all(1)]
```

```python
nrc
```

```python
nrc[nrc[0] == 'happy']
```

```python
nrc[nrc[0] == 'disaster']
```

```python
nrc = nrc.reset_index(drop=True)
nrc
```

```python
lines1 = pos_text
```

```python
stopwords = stopwords.words('english')
```

```python
tokenizer = RegexpTokenizer('[\w]+')
```

```python
p_stemmer = PorterStemmer()
```

```python
raw = lines1.lower()
tokens = tokenizer.tokenize(raw)
stopped_tokens = [i for i in tokens if not i in stopwords]
stopped_tokens
```

```python
match_tokens = [x for x in stopped_tokens if x in list(nrc[0])]
match_tokens               # 감성 사전에 포함돼있던 토큰들
```

```python
emotion = []

for i in match_tokens:
    em_word = nrc[nrc[0] == i][1].values
    
    for j in em_word:
        emotion.append(j)
emotion                              # 문서에 포함돼있던 감성 단어들
```

```python
sent_result = pd.Series(emotion).value_counts()
sent_result
```

```python
def emotion_score(series):
    
    emot_score = 0
    
    for j in range(len(series)):
        if series.index[j] in ('positive', 'anticipation', 'trust', 'joy', 'surprise'):
            emot_score += series.iloc[j]
        else:
            emot_score -= series.iloc[j]
    return emot_score
```

```python
emotion_score(sent_result)
```

```python
lines2 = neg_text

raw = line2.lower()
tokens = tokenizer.tokenize(raw)
stopped_tokens = [i for i in tokens if not i in stopwords]
match_tokens = [x for x in stopped_tokens if x in list(nrc[0])]

emotion = []

for i in match_tokens:
    em_word = nrc[nrc[0] == i][1].values
    
    for j in em_word:
        emotion.append(j)

sent_result = pd.Series(emotion).value_counts()


emotion_score(sent_result)
```

```python
table = dict()

with open('./data_files/polarity.csv', 'r', -1, 'utf-8') as polarity:
    next(polarity)
    
    for line in csv.reader(polarity):
        key = str()
        for word in line[0].split(';'):
            key += word.split('/')[0]
        table[key] = {'neg':line[3], 'neut':line[4], 'pos':line[6]}
        
      
```
